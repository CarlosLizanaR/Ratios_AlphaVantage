```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def get_financial_data(symbol, apikey, endpoint, periods=10):
    """Obtiene los datos financieros de Alpha Vantage según el endpoint indicado (INCOME_STATEMENT o BALANCE_SHEET)."""
    base_url = "https://www.alphavantage.co/query"
    params = {
        "function": endpoint,
        "symbol": symbol,
        "apikey": apikey
    }
    
    response = requests.get(base_url, params=params)
    data = response.json()
    
    key = "annualReports"
    if key in data:
        reports = data[key][:periods]  # Limita a los últimos 'periods' años
        df = pd.DataFrame(reports)
        df.set_index("fiscalDateEnding", inplace=True)
        return df
    else:
        print(f"Error al obtener los datos de {endpoint}: {data}")
        return None

# Solicitar inputs al usuario
symbol = input("Ingrese el símbolo de la empresa (ej. NVDA, AAPL, MSFT): ").upper()
periods = int(input("Ingrese el número de períodos (años) que desea analizar (ej. 5, 10): "))

# Parámetros
apikey = "76259FLG0522EP4I"  # Tu API Key de Alpha Vantage

# Obtener Balance Sheet
df_income = get_financial_data(symbol, apikey, "INCOME_STATEMENT", periods)
df_balance = get_financial_data(symbol, apikey, "BALANCE_SHEET", periods)
df_ratios = pd.merge(df_income, df_balance, on="fiscalDateEnding", how="outer")

# Verificar si se obtuvieron datos
if df_balance is not None:
    # Convertir columnas a numéricas
    df_balance['totalLiabilities'] = pd.to_numeric(df_balance['totalLiabilities'], errors='coerce')
    df_balance['totalShareholderEquity'] = pd.to_numeric(df_balance['totalShareholderEquity'], errors='coerce')
    df_balance['totalAssets'] = pd.to_numeric(df_balance['totalAssets'], errors='coerce')
    df_income['totalRevenue'] = pd.to_numeric(df_income['totalRevenue'], errors='coerce')
    df_income['costOfRevenue'] = pd.to_numeric(df_income['costOfRevenue'], errors='coerce')
    df_balance['inventory'] = pd.to_numeric(df_balance['inventory'], errors='coerce')
    df_income['netIncome'] = pd.to_numeric(df_income['netIncome'], errors='coerce')
    df_income['grossProfit'] = pd.to_numeric(df_income['grossProfit'], errors='coerce')
    df_balance['commonStockSharesOutstanding'] = pd.to_numeric(df_balance['commonStockSharesOutstanding'], errors='coerce')
    df_balance['totalCurrentAssets'] = pd.to_numeric(df_balance['totalCurrentAssets'], errors='coerce')
    df_balance['totalCurrentLiabilities'] = pd.to_numeric(df_balance['totalCurrentLiabilities'], errors='coerce')
    df_income['interestAndDebtExpense'] = pd.to_numeric(df_income['interestAndDebtExpense'], errors='coerce')
    df_income['ebit'] = pd.to_numeric(df_income['ebit'], errors='coerce')
    df_income['ebitda'] = pd.to_numeric(df_income['ebitda'], errors='coerce')

    # Calcular el Ratio de Apalancamiento
    df_ratios['Ratio de Apalancamiento'] = df_balance['totalLiabilities'] / df_balance['totalShareholderEquity']

    # Calcular el Ratio de Endeudamiento
    df_ratios['Ratio de Endeudamiento'] = df_balance['totalLiabilities'] / df_balance['totalAssets']

    # Calcular el Ratio de Rotación de Activos
    df_ratios['Rotación de Activos'] = df_income['totalRevenue'] / df_balance['totalAssets']

    # Calcular el Ratio de Rotación de Inventarios
    df_ratios['Rotación de Inventarios'] = df_income['costOfRevenue'] / df_balance['inventory']

    # Calcular el ROA
    df_ratios['ROA'] = df_income['netIncome'] / df_balance['totalAssets']

    # Calcular el ROE
    df_ratios['ROE'] = df_income['netIncome'] / df_balance['totalShareholderEquity']

    # Calcular el Margen Bruto
    df_ratios['Margen Bruto'] = df_income['grossProfit'] / df_income['totalRevenue']

    # Calcular el Margen Neto
    df_ratios['Margen Neto'] = df_income['netIncome'] / df_income['totalRevenue']

    # Calcular el Beneficio por acción
    df_ratios['Beneficio por acción'] = df_income['netIncome'] / df_balance['commonStockSharesOutstanding']

    # Obtener los años desde los índices
    años = [int(str(fecha).split('-')[0]) for fecha in df_ratios.index]
    bpa = df_ratios['Beneficio por acción'].values

    # Ajustar una recta de tendencia
    coef = np.polyfit(años, bpa, 1)
    tendencia = np.poly1d(coef)

    # Calcular el Ratio de liquidez
    df_ratios['Ratio de liquidez'] = df_balance['totalCurrentAssets'] / df_balance['totalCurrentLiabilities']

    # Calcular el Test Acido
    df_ratios['Test acido'] = (df_balance['totalCurrentAssets'] - df_balance['inventory']) / df_balance['totalCurrentLiabilities']

    # Calcular la Cobertura de Intereses con EBIT
    df_ratios['Cobertura de Intereses con EBIT'] = df_income['ebit'] / df_income['interestAndDebtExpense']

    # Calcular la Cobertura de Intereses con EBITDA
    df_ratios['Cobertura de Intereses con EBITDA'] = df_income['ebitda'] / df_income['interestAndDebtExpense']

    # Calcular los promedios
    promedio_apalancamiento = df_ratios['Ratio de Apalancamiento'].mean()
    promedio_endeudamiento = df_ratios['Ratio de Endeudamiento'].mean()
    promedio_rotacion = df_ratios['Rotación de Activos'].mean()
    promedio_inventarios = df_ratios['Rotación de Inventarios'].mean()
    promedio_ROA = df_ratios['ROA'].mean()
    promedio_ROE = df_ratios['ROE'].mean()
    promedio_MargenBruto = df_ratios['Margen Bruto'].mean()
    promedio_MargenNeto = df_ratios['Margen Neto'].mean()
    promedio_BPA = df_ratios['Beneficio por acción'].mean()
    promedio_liquidez = df_ratios['Ratio de liquidez'].mean()
    promedio_acido = df_ratios['Test acido'].mean()
    promedio_coberturaebit = df_ratios['Cobertura de Intereses con EBIT'].mean()
    promedio_coberturaebitda = df_ratios['Cobertura de Intereses con EBITDA'].mean()

    # Determinar la alerta del Ratio de Apalancamiento
    if promedio_apalancamiento <= 0.55:
        alerta_apalancamiento = "🟢 Bajo (<= 0.55): Indica que la empresa tiene una sólida base de capital y no depende en exceso de pasivos corrientes para financiar su operación. Es positivo."
    elif 0.55 < promedio_apalancamiento < 1.0:
        alerta_apalancamiento = "🟡 Moderado (>0.55 < 1.0): Aceptable, pero muestra cierta dependencia de pasivos corrientes."
    else:
        alerta_apalancamiento = "🔴 Riesgoso: Indica que los pasivos corrientes superan al patrimonio, lo que puede comprometer la estabilidad financiera si no hay suficiente liquidez."

    # Determinar la alerta del Ratio de Endeudamiento
    if promedio_endeudamiento <= 0.33:
        alerta_endeudamiento = "🟢 Bajo (<= 0.33): Excelente solvencia, bajo riesgo financiero. La empresa opera mayormente con capital propio."
    elif 0.33 < promedio_endeudamiento < 0.66:
        alerta_endeudamiento = "🟡 Moderado (>0.33 < 0.66): Nivel de endeudamiento aceptable, la empresa tiene equilibrio entre deuda y capital propio."
    else:
        alerta_endeudamiento = "🔴 Riesgoso: La empresa depende en gran medida de financiamiento externo, lo que la hace más vulnerable a crisis económicas o alzas en tasas de interés."

    # Determinar la alerta de Rotación de Activos
    if promedio_rotacion >= 0.9:
        alerta_rotacion = "🟢 Alto (> 0.9): Eficiencia operativa sobresaliente. La empresa genera muchos ingresos con pocos activos. Sin embargo, es necesario verificar que la alta eficiencia no esté sacrificando el potencial de crecimiento futuro."
    elif 0.66 <= promedio_rotacion < 0.9:
        alerta_rotacion = "🟡 Moderado (>0.66 < 0.9): La empresa mantiene un equilibrio razonable entre la generación de ingresos y el uso de activos. No es la más eficiente, pero está en una posición aceptable."
    else:
        alerta_rotacion = "🔴 Bajo (< 0.66): La empresa está subutilizando sus activos, lo que puede ser un indicador de ineficiencia operativa o sobreinversión en activos. Esto podría requerir una revisión estratégica de cómo se gestionan los recursos."

    # Determinar la alerta de Rotación de Inventarios
    if promedio_inventarios > 6:
        alerta_inventarios = "🟢 Alto (> 6): Excelente gestión de inventarios. La empresa tiene un buen flujo de ventas y una estrategia eficaz de reposición. No hay exceso de inventario, lo que reduce costos de almacenamiento y obsolescencia. Sin embargo, se debe tener cuidado de no quedarse sin productos disponibles."
    elif 3 <= promedio_inventarios < 6:
        alerta_inventarios = "🟡 Moderado (>3 < 6): Buen equilibrio entre ventas e inventarios. La empresa está gestionando su inventario de manera eficiente sin generar grandes costos adicionales. Es un nivel aceptable para muchas industrias."
    else:
        alerta_inventarios = "🔴 Bajo (< 0.66): Indica ineficiencia en la gestión de inventarios. La empresa tiene exceso de inventarios o no está generando ventas suficientes. Esto puede generar costos adicionales y afectar la rentabilidad, y puede requerir una revisión en la estrategia de compras y ventas."

    # Determinar la alerta de ROA
    if promedio_ROA > 0.09:
        alerta_ROA = "🟢 Alto (> 0.09): Indica fuerte rentabilidad y eficiencia en el uso de activos."
    elif 0.033 <= promedio_ROA < 0.09:
        alerta_ROA = "🟡 Moderado (>0.09 < 0.033): Sostenible y equilibrado, pero con margen de mejora."
    else:
        alerta_ROA = "🔴 Bajo (< 0.033): Puede indicar ineficiencia o problemas en la rentabilidad."

    # Determinar la alerta de ROE
    if promedio_ROE > 0.15:
        alerta_ROE = "🟢 Alto (> 0.15): Puede ser señal de ventaja competitiva y uso inteligente del apalancamiento."
    elif 0.08 < promedio_ROE <= 0.15:
        alerta_ROE = "🟡 Moderado (>0.15 < 0.08): Puede ser común en empresas con estrategias de crecimiento estable y un uso conservador del apalancamiento."
    else:
        alerta_ROE = "🔴 Bajo (< 0.08): Puede estar afectado por baja rentabilidad neta, baja rotación de activos o apalancamiento insuficiente."

    # Determinar la alerta de Margen Bruto
    if promedio_MargenBruto > 0.6:
        alerta_MargenBruto = "🟢 Alto (> 0.6): Indica alta rentabilidad, ventaja competitiva y capacidad de fijación de precios."
    elif 0.35 < promedio_MargenBruto <= 0.6:
        alerta_MargenBruto = "🟡 Moderado (<0.6 > 0.35): Muestra rentabilidad estable y sostenida, con buen control de costos."
    else:
        alerta_MargenBruto = "🔴 Bajo (< 0.35): Puede señalar dificultades competitivas o costos elevados, afectando la rentabilidad."

    # Determinar la alerta de Margen Neto
    if promedio_MargenNeto > 0.21:
        alerta_MargenNeto = "🟢 Alto (> 0.21): Indica buena eficiencia operativa, control de costos y posible ventaja competitiva. La empresa es altamente rentable después de cubrir todos sus gastos."
    elif 0.07 < promedio_MargenNeto <= 0.21:
        alerta_MargenNeto = "🟡 Moderado (>0.21 < 0.07): Muestra rentabilidad aceptable y sostenida, aunque con áreas de optimización."
    else:
        alerta_MargenNeto = "🔴 Bajo (< 0.07): Puede reflejar problemas financieros, altos costos o presión competitiva."

    # Determinar la tendencia
    pendiente = coef[0]
    if pendiente > 0:
        tendencia_texto = "🟢 Alcista"
        color_tendencia = 'green'
    elif pendiente < 0:
        tendencia_texto = "🔴 Bajista"
        color_tendencia = 'red'
    else:
        tendencia_texto = "🟡 Lateral"
        color_tendencia = 'blue'

    # Determinar la alerta de Ratio de liquidez
    if promedio_liquidez > 2:
        alerta_liquidez = "🟡 Alto (> 2): Sólida capacidad de pago, pero podría reflejar ineficiencia en la gestión del capital."
    elif 1 < promedio_liquidez <= 2:
        alerta_liquidez = "🟢 Moderado (>2 < 1): Equilibrio financiero óptimo, señal de buena gestión."
    else:
        alerta_liquidez = "🔴 Bajo (< 1): Posible riesgo de liquidez, requiere análisis detallado del flujo de caja y financiamiento."

    # Determinar la alerta de Test ácido
    if promedio_acido > 1:
        alerta_acido = "🟢 Alto (> 1): Sólida capacidad de pago sin depender del inventario, ideal para empresas con buena gestión financiera."
    elif 0.7 < promedio_acido <= 1:
        alerta_acido = "🟡 Moderado (>1 < 0.7): Equilibrio razonable, pero con ligera dependencia del inventario o del financiamiento externo."
    else:
        alerta_acido = "🔴 Bajo (< 0.7): Posible riesgo de liquidez, la empresa podría enfrentar dificultades para pagar sus deudas sin vender inventario rápidamente."

    # Determinar la alerta de Cobertura de Intereses con EBIT
    if promedio_coberturaebit > 5:
        alerta_coberturaebit = "🟢 Alto (> 5): Muy fuerte, indica alta rentabilidad operativa y bajo riesgo financiero."
    elif 1.5 < promedio_coberturaebit < 5:
        alerta_coberturaebit = "🟡 Moderado (>1.5 < 5): Riesgo moderado, la empresa cubre sus intereses pero con margen limitado."
    else:
        alerta_coberturaebit = "🔴 Bajo (< 1.5): Riesgo financiero alto, puede tener problemas para pagar intereses."

    # Determinar la alerta de Cobertura de Intereses con EBITDA
    if promedio_coberturaebitda > 6:
        alerta_coberturaebitda = "🟢 Alto (> 6): Excelente, empresa con flujo operativo fuerte y bajo riesgo financiero."
    elif 2 < promedio_coberturaebitda <= 6:
        alerta_coberturaebitda = "🟡 Moderado (>2 < 6): Moderado, puede depender del sector."
    else:
        alerta_coberturaebitda = "🔴 Bajo (< 2): Riesgo alto, bajo margen de seguridad."

    # Graficar ambos Ratios
    plt.figure(figsize=(10, 5))
    
    # Gráfico Ratio de Apalancamiento
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(df_ratios.index, df_ratios['Ratio de Apalancamiento'], marker='o', linestyle='-', label="Ratio de Apalancamiento")
    plt.axhline(y=promedio_apalancamiento, color='r', linestyle='--', label="Promedio")
    plt.xlabel("Periodo")
    plt.ylabel("Ratio de Apalancamiento")
    plt.title(f"Evolución del Ratio de Apalancamiento ({symbol})")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()

    # Gráfico Ratio de Endeudamiento
    plt.subplot(1, 2, 2)
    plt.plot(df_ratios.index, df_ratios['Ratio de Endeudamiento'], marker='s', linestyle='-', label="Ratio de Endeudamiento", color='g')
    plt.axhline(y=promedio_endeudamiento, color='r', linestyle='--', label="Promedio")
    plt.xlabel("Periodo")
    plt.ylabel("Ratio de Endeudamiento")
    plt.title("Evolución del Ratio de Endeudamiento")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()

    # Imprimir resultados
    print("\nRatios Financieros:")
    print("Solvencia")

    # Apalancamiento
    print("\nRatio de Apalancamiento:")
    print("Mide qué proporción de la financiación de corto plazo (pasivos corrientes) está respaldada por los recursos propios de los accionistas.\n")
    print(df_ratios[['Ratio de Apalancamiento']])
    print("\nPromedio del Ratio de Apalancamiento:", promedio_apalancamiento)
    print(f"Alerta basada en el promedio: {alerta_apalancamiento}")

        # Endeudamiento
    print("\nRatio de Endeudamiento:")
    print("Mide qué proporción de los activos de la empresa está financiada con deuda en lugar de capital propio.\n")
    print(df_ratios[['Ratio de Endeudamiento']])
    print("\nPromedio del Ratio de Endeudamiento:", promedio_endeudamiento)
    print(f"Alerta basada en el promedio: {alerta_endeudamiento}")

    # Graficar ambos Ratios
    plt.figure(figsize=(10, 5))      

    # Crear gráfico de la Rotación de Activos
    plt.subplot(1, 2, 1)
    plt.plot(df_ratios.index, df_ratios['Rotación de Activos'], marker='o', linestyle='-', label="Rotación de Activos")
    plt.axhline(y=promedio_rotacion, color='r', linestyle='--', label="Promedio")
    plt.xlabel("Periodo")
    plt.ylabel("Rotación de Activos")
    plt.title("Evolución de Rotación de Activos")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()

    # Crear gráfico de la Rotación de Inventarios
    plt.subplot(1, 2, 2)
    plt.plot(df_ratios.index, df_ratios['Rotación de Inventarios'], marker='s', linestyle='-', label="Rotación de Inventarios", color='green')
    plt.axhline(y=promedio_inventarios, color='r', linestyle='--', label="Promedio")
    plt.xlabel("Periodo")
    plt.ylabel("Rotación de Inventarios")
    plt.title("Evolución de la Rotación de Inventarios")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()
    
    plt.tight_layout()
    plt.show()

    # Imprimir resultados
    print("\nRatios Financieros:")
    print("Rotación")
    print("Rotación de Activos")
    print("La rotación de activos es un ratio de eficiencia que mide la capacidad de una empresa para generar ingresos a partir de sus activos totales. Este ratio indica cuántos dólares de ventas son generados por cada unidad monetaria de activos invertidos en el negocio. Es una medida de cuán eficientemente la empresa está utilizando sus activos para generar ingresos.\n")
    print(df_ratios[['Rotación de Activos']])
    print("\nPromedio del Ratio de Rotación de Activos:", promedio_rotacion)
    print(f"Alerta basada en el promedio: {alerta_rotacion}")    

    print("Rotación de Inventarios")
    print("La rotación de inventarios es un ratio financiero que mide la eficiencia con la que una empresa vende y repone su inventario durante un período determinado. Indica cuántas veces se ha renovado el inventario a lo largo de un año, reflejando la rapidez con que los productos se venden y se reabastecen.\n")
    print(df_ratios[['Rotación de Inventarios']])
    print("\nPromedio de la Rotación de Inventarios:", promedio_inventarios)
    print(f"Alerta basada en el promedio: {alerta_inventarios}")

    # Graficar ambos Ratios
    plt.figure(figsize=(10, 5))  

    # Crear gráfico del ROA
    plt.subplot(1, 2, 1)
    plt.plot(df_ratios.index, df_ratios['ROA'], marker='o', linestyle='-', label="ROA")
    plt.axhline(y=promedio_ROA, color='r', linestyle='--', label="Promedio")
    plt.xlabel("Periodo")
    plt.ylabel("ROA")
    plt.title("Evolución del ROA")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()

    # Crear gráfico del ROE
    plt.subplot(1, 2, 2)
    plt.plot(df_ratios.index, df_ratios['ROE'], marker='s', linestyle='-', label="ROE", color='green')
    plt.axhline(y=promedio_ROE, color='r', linestyle='--', label="Promedio")
    plt.xlabel("Periodo")
    plt.ylabel("ROE")
    plt.title("Evolución del ROE")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()

    # Imprimir resultados
    print("\nRatios Financieros:")
    print("Rentabilidad")
    print("ROA")
    print("El ROA (Return on Assets o Rentabilidad sobre Activos) es un indicador financiero que mide la eficiencia con la que una empresa utiliza sus activos para generar ganancias.\n")
    print(df_ratios[['ROA']])
    print("\nPromedio del Ratio del ROA:", promedio_ROA)
    print(f"Alerta basada en el promedio: {alerta_ROA}")

    print("ROE")
    print("El ROE (Return on Equity o Rentabilidad sobre el Patrimonio) es un indicador financiero que mide la rentabilidad que obtiene una empresa en relación con el capital aportado por los accionistas.\n")
    print(df_ratios[['ROE']])
    print("\nPromedio del Ratio de Rotación del ROE:", promedio_ROE)
    print(f"Alerta basada en el promedio: {alerta_ROE}")


    # Graficar ambos Ratios
    plt.figure(figsize=(10, 5))

    # Crear gráfico del Margen Bruto
    plt.subplot(1, 2, 1)
    plt.plot(df_ratios.index, df_ratios['Margen Bruto'], marker='o', linestyle='-', label="ROA")
    plt.axhline(y=promedio_MargenBruto, color='r', linestyle='--', label="Promedio")
    plt.xlabel("Periodo")
    plt.ylabel("Margen bruto")
    plt.title("Evolución del Margen bruto")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()

    # Crear gráfico del Margen Neto
    plt.subplot(1, 2, 2)
    plt.plot(df_ratios.index, df_ratios['Margen Neto'], marker='s', linestyle='-', label="ROE", color='green')
    plt.axhline(y=promedio_MargenNeto, color='r', linestyle='--', label="Promedio")
    plt.xlabel("Periodo")
    plt.ylabel("Margen Neto")
    plt.title("Evolución del Margen Neto")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()

    print("Margen Bruto:")
    print("El margen bruto es un indicador financiero que mide la rentabilidad de una empresa a nivel operativo, es decir, cuánto gana después de cubrir los costos directos de producción. \n")
    print(df_ratios[['Margen Bruto']])
    print("\nPromedio del Margen Bruto:", promedio_MargenBruto)
    print(f"Alerta basada en el promedio: {alerta_MargenBruto}")

    print("Margen Neto:")
    print("El margen neto es un indicador de rentabilidad que mide el porcentaje de ingresos que una empresa convierte en utilidad neta después de deducir todos los costos y gastos, incluidos los operativos, financieros, impuestos y otros gastos no operacionales. \n")
    print(df_ratios[['Margen Neto']])
    print("\nPromedio del Margen Neto:", promedio_MargenNeto)
    print(f"Alerta basada en el promedio: {alerta_MargenNeto}")

    # Crear gráfico
    fig, ax1 = plt.subplots(figsize=(10, 5))
    
    # Eje izquierdo: BPA real
    ax1.set_xlabel("Año")
    ax1.set_ylabel("Beneficio por Acción", color='tab:blue')
    ax1.plot(años, bpa, marker='o', linestyle='-', color='tab:blue', label="BPA Real")
    ax1.tick_params(axis='y', labelcolor='tab:blue')
    
    # Eje derecho: Tendencia del BPA
    ax2 = ax1.twinx()
    ax2.set_ylabel("Tendencia BPA", color='tab:red')
    ax2.plot(años, tendencia(años), linestyle='--', color='tab:red', label="Tendencia BPA")
    ax2.tick_params(axis='y', labelcolor='tab:red')
    
    # Títulos y leyenda
    fig.suptitle("Evolución del Beneficio por Acción (BPA) y su Tendencia")
    fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.9))
    
    plt.show()

    print("Beneficio por acción:")
    print("El Beneficio por Acción (BPA) es un indicador financiero que muestra la ganancia neta que corresponde a cada acción en circulación de una empresa. Es fundamental para evaluar la rentabilidad de una compañía desde la perspectiva del accionista. \n")
    print(df_ratios[['Beneficio por acción']])
    print("\nPromedio de BPA:", promedio_BPA)
    print(f"Alerta basada en la tendencia: {tendencia_texto}")

    # Graficar ambos Ratios
    plt.figure(figsize=(10, 5))

    # Crear gráfico de la Rotación de Ratio de liquidez
    plt.subplot(1, 2, 1)
    plt.plot(df_ratios.index, df_ratios['Ratio de liquidez'], marker='o', linestyle='-', label="Ratio de liquidez")
    plt.axhline(y=promedio_liquidez, color='r', linestyle='--', label="Promedio")
    plt.xlabel("Periodo")
    plt.ylabel("Ratio de liquidez")
    plt.title("Evolución del Ratio de liquidez")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()

    # Crear gráfico del Test acido
    plt.subplot(1, 2, 2)
    plt.plot(df_ratios.index, df_ratios['Test acido'], marker='s', linestyle='-', label="Test acido", color='green')
    plt.axhline(y=promedio_acido, color='r', linestyle='--', label="Promedio")
    plt.xlabel("Periodo")
    plt.ylabel("Test acido")
    plt.title("Evolución del Test acido")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()
    
    plt.tight_layout()
    plt.show()

    # Imprimir resultados
    print("Liquidez")
    print("Ratio de liquidez")
    print("El ratio de liquidez mide la capacidad de una empresa para cubrir sus obligaciones de corto plazo con sus activos circulantes.\n")
    print(df_ratios[['Ratio de liquidez']])
    print("\nPromedio del Ratio de liquidez:", promedio_liquidez)
    print(f"Alerta basada en el promedio: {alerta_liquidez}")

    print("Test acido")
    print("El Test Ácido mide la capacidad de una empresa para cubrir sus pasivos circulantes sin depender del inventario. Es una versión más estricta del Ratio de Liquidez, ya que excluye los activos menos líquidos (como el inventario), enfocándose en los recursos que pueden convertirse rápidamente en efectivo. \n")
    print(df_ratios[['Test acido']])
    print("\nPromedio del Test acido:", promedio_acido)
    print(f"Alerta basada en el promedio: {alerta_acido}")

    # Graficar ambos Ratios
    plt.figure(figsize=(10, 5))

    # Crear gráfico de la Cobertura de Intereses con EBIT
    plt.subplot(1, 2, 1)
    plt.plot(df_ratios.index, df_ratios['Cobertura de Intereses con EBIT'], marker='o', linestyle='-', label="Cobertura de Intereses con EBIT")
    plt.axhline(y=promedio_coberturaebit, color='r', linestyle='--', label="Promedio")
    plt.xlabel("Periodo")
    plt.ylabel("Cobertura de Intereses con EBIT")
    plt.title("Evolución de Cobertura de Intereses con EBIT")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()

    # Crear gráfico de la Cobertura de Intereses con EBITDA
    plt.subplot(1, 2, 2)
    plt.plot(df_ratios.index, df_ratios['Cobertura de Intereses con EBITDA'], marker='o', linestyle='-', label="Cobertura de Intereses con EBITDA")
    plt.axhline(y=promedio_coberturaebitda, color='r', linestyle='--', label="Promedio")
    plt.xlabel("Periodo")
    plt.ylabel("Cobertura de Intereses con EBITDA")
    plt.title("Evolución de Cobertura de Intereses con EBITDA")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()

    # Imprimir resultados
    print("Cobertura")
    print("Cobertura de Intereses con EBIT")
    print("Mide cuántas veces la empresa puede pagar sus intereses con su ganancia operativa. Un valor alto indica mayor capacidad para cubrir la deuda sin afectar su operación.\n")
    print(df_ratios[['Cobertura de Intereses con EBIT']])
    print("\nPromedio de Cobertura de Intereses con EBIT", promedio_coberturaebit)
    print(f"Alerta basada en el promedio: {alerta_coberturaebit}")

    print("Cobertura de Intereses con EBITDA")
    print(" Evalúa la capacidad de pago de intereses considerando el flujo operativo antes de depreciaciones y amortizaciones. Es útil para comparar empresas con altos costos no monetarios.\n")
    print(df_ratios[['Cobertura de Intereses con EBITDA']])
    print("\nPromedio de Cobertura de Intereses con EBITDA", promedio_coberturaebitda)
    print(f"Alerta basada en el promedio: {alerta_coberturaebitda}")


else:
    print(f"No se pudieron obtener datos para el símbolo {symbol}. Verifica el símbolo e intenta nuevamente.")
```
